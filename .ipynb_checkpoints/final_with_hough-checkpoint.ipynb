```python
def isolate_traffic_sign_colors_strict(image):
    """
    Strictly isolate ONLY official traffic sign colors with narrow ranges
    - Bright red (warning/prohibition signs)
    - DEEP blue ONLY (mandatory/information signs) - NO sky blue, NO soft blue
    - Bright yellow/orange (warning signs)
    
    Excludes: green, cyan, pale colors, dark colors, sky blue, light blue, WHITE
    """
    
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # STRICT RED - Only bright, saturated red (no pink, no brown)
    lower_red1 = np.array([0, 120, 100])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 120, 100])
    upper_red2 = np.array([180, 255, 255])
    
    # VERY STRICT DEEP BLUE - Only traffic sign deep blue
    lower_blue = np.array([110, 150, 100])
    upper_blue = np.array([120, 255, 220])
    
    # STRICT YELLOW - Only bright yellow/orange
    lower_yellow = np.array([18, 120, 150])
    upper_yellow = np.array([32, 255, 255])
    
    # Create masks for each color (NO WHITE)
    mask_red1 = cv2.inRange(hsv, lower_red1, upper_red1)
    mask_red2 = cv2.inRange(hsv, lower_red2, upper_red2)
    mask_red = cv2.bitwise_or(mask_red1, mask_red2)
    
    mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
    mask_yellow = cv2.inRange(hsv, lower_yellow, upper_yellow)
    
    # Combine masks (RED, BLUE, YELLOW only)
    combined_mask = cv2.bitwise_or(mask_red, mask_blue)
    combined_mask = cv2.bitwise_or(combined_mask, mask_yellow)
    
    # Aggressive noise removal
    kernel = np.ones((3, 3), np.uint8)
    combined_mask = cv2.morphologyEx(combined_mask, cv2.MORPH_CLOSE, kernel, iterations=2)
    combined_mask = cv2.morphologyEx(combined_mask, cv2.MORPH_OPEN, kernel, iterations=2)
    
    return combined_mask
```

```python
def gaussian_blur_canny(image, kernel_size=5, canny_low=50, canny_high=150): 

        # Step 1: Strict color isolation
    color_mask = isolate_traffic_sign_colors_strict(image)

    plt.imshow(color_mask,cmap='gray')
    plt.axis('off')
    plt.show()
    
    # Step 2: Apply mask
    isolated = cv2.bitwise_and(image, image, mask=color_mask)
   
    gray = cv2.cvtColor(isolated, cv2.COLOR_BGR2GRAY)
    
    
    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0)
    
    # Apply Canny edge detection
    edges = cv2.Canny(blurred, canny_low, canny_high)
    
    return edges,color_mask
```

```python
canny,color_mask=gaussian_blur_canny(img)
plt.imshow(canny,cmap='gray')
plt.axis('off')
plt.show()
```

```python
def extract_shape_boundaries_adaptive(image, kernel_size=5, canny_low=50, canny_high=150, min_area_ratio=0.01):
   
    gray = image.copy()
    
    # Calculate adaptive min_area based on image size
    total_pixels = gray.shape[0] * gray.shape[1]
    min_area = int(total_pixels * min_area_ratio)
    
    print(f"Image size: {gray.shape[0]}x{gray.shape[1]} = {total_pixels} pixels")
    print(f"Min area threshold: {min_area} pixels ({min_area_ratio*100}% of image)")
    
    # Find contours
    contours, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Create blank image for drawing clean boundaries
    result = np.zeros_like(gray)
    
    # Filter and draw only significant shape boundaries
    shapes_found = 0
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > min_area:  # Adaptive filtering
            cv2.drawContours(result, [contour], -1, 255, 1)
            shapes_found += 1
    
    print(f"Found {shapes_found} significant shapes")
    
    return result

def extract_shape_boundaries_filled(image, kernel_size=5, canny_low=50, canny_high=150, min_area_ratio=0.01):
   
    # Get basic boundaries first
    result = extract_shape_boundaries_adaptive(image, kernel_size, canny_low, canny_high, min_area_ratio)
    
    # Fill gaps using morphological closing
    kernel = np.ones((3, 3), np.uint8)
    
    # Close small gaps
    result = cv2.morphologyEx(result, cv2.MORPH_CLOSE, kernel, iterations=1)
    
    # Optional: Make lines slightly thicker
    #result = cv2.dilate(result, kernel, iterations=1)
    
    return result

```

```python
def extract_inner_region(original_image, boundary_image):

    # Create a mask by filling the boundaries
    mask = boundary_image.copy()
    
    # Find contours from the boundary
    contours, _ = cv2.findContours(boundary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Fill the interior of each contour to create a solid mask
    for contour in contours:
        cv2.fillPoly(mask, [contour], 255)
    
    # Normalize mask to 0-1 range for proper masking
    mask_normalized = mask.astype(np.float32) / 255.0
    
    # Apply the mask to preserve original colors
    if len(original_image.shape) == 3:  # Color image
        # Apply mask to each channel separately
        result = original_image.copy().astype(np.float32)
        for i in range(3):
            result[:, :, i] = result[:, :, i] * mask_normalized
        result = result.astype(np.uint8)
    else:  # Grayscale image
        result = (original_image.astype(np.float32) * mask_normalized).astype(np.uint8)
    
    return result, mask

def extract_inner_region_cropped(original_image, boundary_image):

    # Get the inner region first
    inner_image, mask = extract_inner_region(original_image, boundary_image)
    
    # Find the bounding box of the shape
    contours, _ = cv2.findContours(boundary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if len(contours) > 0:
        # Get the largest contour (main shape)
        largest_contour = max(contours, key=cv2.contourArea)
        
        # Get bounding rectangle
        x, y, w, h = cv2.boundingRect(largest_contour)
        
        # Crop to the bounding box
        cropped = inner_image[y:y+h, x:x+w]
        
        return cropped, (x, y, w, h)
    
    return inner_image, None
```

```python
def analyze_colors(image):
    # Convert to HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # Define HSV ranges for ONLY significant road sign colors
    color_ranges = {
        'red': [
            (np.array([0, 100, 100]), np.array([10, 255, 255])),    # Stop, warning, prohibition signs
            (np.array([170, 100, 100]), np.array([180, 255, 255]))  # Stop, warning, prohibition signs
        ],
        'blue': [(np.array([100, 100, 100]), np.array([130, 255, 255]))],  # Mandatory/information signs
        'yellow': [(np.array([20, 100, 100]), np.array([30, 255, 255]))],   # Warning signs
        'green': [(np.array([40, 100, 100]), np.array([80, 255, 255]))],    # Direction/guide signs
        'orange': [(np.array([10, 100, 100]), np.array([20, 255, 255]))],   # Construction/temporary signs
        'brown': [(np.array([8, 100, 50]), np.array([15, 255, 150]))]       # Tourist/recreational signs
    }
    
    color_pixels = {}
    
    # Only count non-black pixels (ignore background)
    non_black_mask = cv2.inRange(hsv, np.array([0, 0, 1]), np.array([180, 255, 255]))
    total_pixels = cv2.countNonZero(non_black_mask)
    
    if total_pixels == 0:
        return "unknown", {}
    
    for color_name, ranges in color_ranges.items():
        mask = np.zeros(hsv.shape[:2], dtype=np.uint8)
        
        for lower, upper in ranges:
            color_mask = cv2.inRange(hsv, lower, upper)
            mask = cv2.bitwise_or(mask, color_mask)
        
        # Only count pixels that are also non-black
        final_mask = cv2.bitwise_and(mask, non_black_mask)
        pixel_count = cv2.countNonZero(final_mask)
        color_pixels[color_name] = pixel_count / total_pixels if total_pixels > 0 else 0
    
    # Find dominant significant color (ignore colors with <2% presence)
    significant_colors = {k: v for k, v in color_pixels.items() if v > 0.02}
    
    if significant_colors:
        dominant_color = max(significant_colors, key=significant_colors.get)
        return dominant_color, significant_colors
    else:
        return "no_significant_color", color_pixels
```

```python
dominant_color, color_percentages = analyze_colors(cropped_inner)
print(f"Dominant color: {dominant_color}")
print("Significant color percentages:")
for color, percentage in sorted(color_percentages.items(), key=lambda x: x[1], reverse=True):
    if percentage > 0.01:  
        print(f"  {color}: {percentage*100:.1f}%")
```

```python
def fill_shape_interior(boundary_image):
    """
    Fill the interior of shapes while preserving geometry
    """
    # Use SMALLER kernel and fewer iterations to preserve shape
    kernel = np.ones((3, 3), np.uint8)
    
    # Just enough to close small gaps, not change geometry
    closed = cv2.morphologyEx(boundary_image, cv2.MORPH_CLOSE, kernel, iterations=2)
    
    # Find contours
    contours, _ = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Create filled version
    filled = np.zeros_like(boundary_image)
    
    # Fill each external contour
    for contour in contours:
        cv2.fillPoly(filled, [contour], 255)
    
    return filled
```

```python
def classify_shape_improved_robust(image):
    """
    Improved shape classification with multiple epsilon testing
    to accurately detect vertex count, especially for triangles
    """
    # Fill shape interior
    filled_image = fill_shape_interior(image)
    
    # Find contours
    contours, _ = cv2.findContours(filled_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if not contours:
        return "unknown", None
    
    # Get the largest contour
    cnt = max(contours, key=cv2.contourArea)
    
    if len(cnt) < 5:
        return "unknown", None
    
    # Use convex hull to get ideal shape
    hull = cv2.convexHull(cnt)
    
    # ============================================
    # MULTI-EPSILON TESTING (Key Improvement!)
    # ============================================
    
    # Test multiple epsilon values to find most stable vertex count
    epsilon_values = [0.01, 0.015, 0.02, 0.025, 0.03, 0.035, 0.04]
    vertex_counts = []
    approximations = []
    
    print("Testing multiple epsilon values:")
    for eps in epsilon_values:
        epsilon = eps * cv2.arcLength(hull, True)
        approx = cv2.approxPolyDP(hull, epsilon, True)
        vertex_count = len(approx)
        vertex_counts.append(vertex_count)
        approximations.append(approx)
        print(f"  ={eps:.3f}  {vertex_count} vertices")
    
    # Strategy 1: Find the most common (stable) vertex count
    from collections import Counter
    count_freq = Counter(vertex_counts)
    most_common_count = count_freq.most_common(1)[0][0]
    print(f"Most common vertex count: {most_common_count} (appears {count_freq[most_common_count]} times)")
    
    # Strategy 2: For low vertex counts (3-8), prefer the value that appears most
    # For high counts (>8), it's likely a circle
    if most_common_count <= 8:
        # Use the approximation from middle epsilon that gives this count
        vertices = most_common_count
        idx = vertex_counts.index(most_common_count)
        approx = approximations[idx]
    else:
        # High vertex count, likely a circle
        vertices = most_common_count
        approx = approximations[len(approximations)//2]  # Use middle epsilon
    
    print(f"Selected: {vertices} vertices")
    
    # ============================================
    # ADDITIONAL VERIFICATION FOR TRIANGLES
    # ============================================
    
    # If we got 4 vertices but shape might be triangle, do angle analysis
    if vertices == 4:
        print("\nVerifying if 4 vertices is actually a triangle...")
        
        # Check if one vertex is very close to the line of two others
        triangle_score = check_if_triangle(approx)
        
        if triangle_score > 0.8:  # High confidence it's a triangle
            print("   Confirmed: This is actually a TRIANGLE (one vertex is spurious)")
            vertices = 3
            # Remove the weakest vertex
            approx = reduce_to_triangle(approx)
    
    # ============================================
    # Calculate geometric properties
    # ============================================
    
    x, y, w, h = cv2.boundingRect(hull)
    aspect_ratio = float(w) / h
    hull_area = cv2.contourArea(hull)
    rect_area = w * h
    extent = float(hull_area) / rect_area
    
    cnt_area = cv2.contourArea(cnt)
    solidity = float(cnt_area) / hull_area if hull_area > 0 else 0
    
    perimeter = cv2.arcLength(hull, True)
    circularity = 4 * np.pi * hull_area / (perimeter**2) if perimeter > 0 else 0
    
    # Eccentricity
    if len(hull) >= 5:
        try:
            (x_center, y_center), (MA, ma), angle = cv2.fitEllipse(hull)
            major_axis = max(MA, ma)
            minor_axis = min(MA, ma)
            
            if major_axis > 0:
                ratio = minor_axis**2 / major_axis**2
                ratio = max(0, min(1, ratio))
                eccentricity = np.sqrt(1 - ratio)
            else:
                eccentricity = 0
        except:
            eccentricity = 0
    else:
        eccentricity = 0
    
    # ============================================
    # SHAPE CLASSIFICATION
    # ============================================
    
    shape_name = "unknown"
    confidence = 0
    
    # Triangle (MUST have exactly 3 vertices)
    if vertices == 3 and solidity > 0.75:
        shape_name = "triangle"
        confidence = solidity * 0.95
        print(" Classified as TRIANGLE")
    
    # Circle (high circularity OR many vertices)
    elif (vertices >= 8 and circularity > 0.85 and eccentricity < 0.05 and 
        solidity > 0.9 and 0.8 < aspect_ratio < 1.2):
        shape_name = "circle"
        confidence = circularity
        print(" Classified as CIRCLE")
    
    # Square (4 vertices + square aspect ratio)
    elif (vertices == 4 and 0.85 < aspect_ratio < 1.15 and 
          extent > 0.75 and solidity > 0.85):
        shape_name = "square"
        confidence = extent * solidity
        print(" Classified as SQUARE")
    
    # Diamond (4 vertices + square aspect + low extent)
    elif (vertices == 4 and 0.85 < aspect_ratio < 1.15 and 
          extent < 0.75 and solidity > 0.85):
        shape_name = "diamond"
        confidence = extent * solidity
        print(" Classified as DIAMOND")
    
    # Rectangle (4 vertices + non-square aspect)
    elif (vertices == 4 and (aspect_ratio < 0.85 or aspect_ratio > 1.15) and
          extent > 0.7 and solidity > 0.80):
        shape_name = "rectangle" 
        confidence = extent * solidity
        print(" Classified as RECTANGLE")
    
    # Pentagon (5 vertices)
    elif vertices == 5 and solidity > 0.80:
        shape_name = "pentagon"
        confidence = solidity * 0.9
        print(" Classified as PENTAGON")
    
    # Hexagon (6 vertices)
    elif vertices == 6 and solidity > 0.80:
        shape_name = "hexagon"
        confidence = solidity * 0.9
        print(" Classified as HEXAGON")
    
    # Octagon (8 vertices + not circular)
    elif vertices == 8 and eccentricity > 0.1 and solidity > 0.80:
        shape_name = "octagon"
        confidence = solidity * 0.9
        print(" Classified as OCTAGON")
    
    # Very high vertex count = circle
    elif vertices > 10 and circularity > 0.6:
        shape_name = "circle"
        confidence = circularity * 0.85
        print(" Classified as CIRCLE (many vertices)")
    
    return shape_name, {
        'confidence': confidence,
        'vertices': vertices,
        'tested_epsilon_values': epsilon_values,
        'vertex_counts_per_epsilon': vertex_counts,
        'aspect_ratio': aspect_ratio,
        'extent': extent,
        'solidity': solidity, 
        'circularity': circularity,
        'eccentricity': eccentricity,
        'area': hull_area,
        'perimeter': perimeter
    }

def check_if_triangle(four_vertices):
    """
    Check if 4 vertices actually form a triangle (one vertex is spurious)
    Returns score 0-1, where 1 means definitely a triangle
    """
    vertices = four_vertices.reshape(-1, 2)
    
    # For each vertex, check if it's close to the line formed by two others
    min_distance_ratio = float('inf')
    
    for i in range(4):
        # Get other vertices
        others_idx = [j for j in range(4) if j != i]
        
        # Check all combinations of two other vertices
        for j in range(len(others_idx)):
            for k in range(j+1, len(others_idx)):
                p1 = vertices[others_idx[j]]
                p2 = vertices[others_idx[k]]
                point = vertices[i]
                
                # Distance from point to line p1-p2
                dist = point_to_line_distance(point, p1, p2)
                
                # Length of line p1-p2
                line_length = np.linalg.norm(p2 - p1)
                
                if line_length > 0:
                    ratio = dist / line_length
                    min_distance_ratio = min(min_distance_ratio, ratio)
    
    # If one vertex is very close to a line (ratio < 0.1), it's likely spurious
    if min_distance_ratio < 0.1:
        return 0.9
    elif min_distance_ratio < 0.15:
        return 0.7
    else:
        return 0.0

def point_to_line_distance(point, line_p1, line_p2):
    """Calculate perpendicular distance from point to line"""
    # Vector from line_p1 to line_p2
    line_vec = line_p2 - line_p1
    line_len = np.linalg.norm(line_vec)
    
    if line_len < 1e-6:
        return np.linalg.norm(point - line_p1)
    
    # Normalize
    line_unitvec = line_vec / line_len
    
    # Vector from line_p1 to point
    point_vec = point - line_p1
    
    # Project
    proj_length = np.dot(point_vec, line_unitvec)
    proj = line_p1 + proj_length * line_unitvec
    
    # Distance
    return np.linalg.norm(point - proj)

def reduce_to_triangle(four_vertices):
    """
    From 4 vertices, remove the one that's most likely spurious
    Returns 3 vertices that best form a triangle
    """
    vertices = four_vertices.reshape(-1, 2)
    
    # Find which vertex is closest to being collinear with two others
    min_distance = float('inf')
    spurious_idx = -1
    
    for i in range(4):
        others_idx = [j for j in range(4) if j != i]
        
        # Find minimum distance to any line formed by two others
        for j in range(len(others_idx)):
            for k in range(j+1, len(others_idx)):
                p1 = vertices[others_idx[j]]
                p2 = vertices[others_idx[k]]
                point = vertices[i]
                
                dist = point_to_line_distance(point, p1, p2)
                
                if dist < min_distance:
                    min_distance = dist
                    spurious_idx = i
    
    # Remove spurious vertex
    triangle_vertices = [four_vertices[j] for j in range(4) if j != spurious_idx]
    return np.array(triangle_vertices)

# Usage:
shape_name, properties = classify_shape_improved_robust(Filled_bound)
print(f"\nFinal Result: {shape_name}")
print(f"Vertices: {properties['vertices']}")
print(f"Confidence: {properties['confidence']:.2f}")
```
